```python
#URL where data is located
url = 'https://apps.bea.gov/regional/zip/CAGDP11.zip'

# Making a directory to dowload my zip file if it doesn't already exist.
folder_name = 'GDP'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    
# Downloading the required zip file from the url and extracting the correct csv into the folder path.
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extract('CAGDP11__ALL_AREAS_2018_2022.csv', path="GDP/")
    
# Creating a dataframe from the extracted CSV.
file = 'GDP/CAGDP11__ALL_AREAS_2018_2022.csv'
df_gdp = pd.read_csv(file, encoding='latin-1')

print(df_gdp.shape[0],'rows of data')
df_gdp.head()
```

### Quality Issue 1:

```python
#visually inspecting a sample of the df_gdp dataframe
df_gdp.sample(5)
```

```python
# Visually inspecting a sample of the df_gdp data frame.
df_ue.sample(5)
```

```python
# Inspecting how many matching unique location names there are between both data sets

# Creating lists of unique location names from both data frames.
gdp_ls = df_gdp['GeoName'].unique()
ue_ls = df_ue['Area_Name'].unique()

# Comparing both lists for matches.
count = 0
for place in gdp_ls:
    if place in ue_ls:
        count += 1
        print(place)
print('There are',count,'matching unique location names between both data sets.')
```

Issue and justification: There is an issue of consistency involving how locations are named between both data sets. When visually inspecting both sets of records df_gdp uses a (county_name, state) naming convention, whereas df_ue uses a (county_name county, state) naming convention. Additionally even though both sets of records use different naming conventions the both still have 68 matching records. There shouldn't be any matching records when using different naming conventions. Finally some of the locations are only the states name which creates major inconsistency in the data.
